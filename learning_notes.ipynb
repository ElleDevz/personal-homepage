### Learning Time: How do I deploy my project on Netlify using GitHub?

**Question:**
How do I deploy my project on Netlify using GitHub?

**Answer:**
1. **Push Your Project to GitHub**
   - Make sure your project is in a git repository and pushed to GitHub.
   - If you haven’t already:
     ```sh
     git init
     git add .
     git commit -m "Initial commit"
     git remote add origin https://github.com/your-username/your-repo-name.git
     git push -u origin main
     ```

2. **Log in to Netlify**
   - Go to https://app.netlify.com/ and log in (or sign up).

3. **Create a New Site**
   - Click “Add new site” → “Import an existing project”.

4. **Connect to GitHub**
   - Authorize Netlify to access your GitHub account (if you haven’t already).
   - Select your repository from the list.

5. **Configure Build Settings**
   - **Build command:** For most React/Next.js/Vite projects, use `npm run build`
   - **Publish directory:** For static sites, use `dist` or `build` or `out` (for Next.js static export)
   - For your Notes App, if it’s React:  
     - Build command: `npm run build`
     - Publish directory: `build`
   - For Next.js with static export:  
     - Build command: `npm run build`
     - Publish directory: `out`

6. **Deploy**
   - Click “Deploy site”.
   - Netlify will clone your repo, install dependencies, build, and deploy automatically.

7. **Get Your Live URL**
   - Once the build finishes, Netlify will give you a live URL (e.g., `https://your-site-name.netlify.app`).

### Learning Time: What Do These Steps Mean for Pushing and Deploying Projects?

**cd /Users/ellec/Downloads/PersonalHomepageProject**  
Change Directory – Moves you into your project’s folder so all git actions happen in the right place.

**git init**  
Initialize Git – Starts a new git repository in your project folder, allowing you to track changes and use version control.

**git remote add origin https://github.com/ElleDevz/LatoyaCarverResume.git**  
Add Remote – Tells git where your project will live online (on GitHub). “origin” is the nickname for this remote location.

**git add .**  
Stage Changes – Prepares all files in your project to be included in the next commit.

**git commit -m "Initial commit: add all project files"**  
Commit – Saves a snapshot of your staged files with a message describing the change.

**git branch -M main**  
Rename Branch – Renames your current branch to “main,” which is the standard default branch for new projects.

**git push -u origin main**  
Push – Uploads your committed changes from your computer to the “main” branch on GitHub.

---

#### Key Definitions
- **Push:** Sending your local project changes to a remote repository (like GitHub) so they’re saved online and can be shared or deployed.
- **Deploy:** Publishing your project to a live server (like Netlify) so anyone can view it on the web.
- **Repository (repo):** A storage space for your project’s files and history, either on your computer or online (like GitHub).
- **Branch:** A parallel version of your project. “main” is the default branch where your main work lives.
- **Commit:** A saved snapshot of your project at a specific point in time, with a message describing what changed.
- **Remote:** The online location (like GitHub) where your project is stored and shared.

These steps and terms are the foundation for collaborating, sharing, and deploying your projects as a developer!

**Q: It's a PDF, how do I transform it into HTML?**

**A:**

1. **Manual Conversion (Best for Simple Documents):**
   - Open your PDF and copy the text and images.
   - Paste them into an HTML file (`index.html`).
   - Manually format the content using HTML tags (`<h1>`, `<p>`, `<img>`, etc.).

2. **Online PDF to HTML Converters:**
   - Use a free online tool like:
     - [pdf2html.com](https://pdf2html.com/)
     - [Zamzar PDF to HTML](https://www.zamzar.com/convert/pdf-to-html/)
   - Upload your PDF, convert, and download the HTML file.
   - Review and clean up the HTML as needed.

3. **Use a Desktop Tool or Command Line:**
   - For automation or more control, use a tool like `pdf2htmlEX`:
     ```sh
     brew install pdf2htmlex
     pdf2htmlEX yourfile.pdf
     ```
   - This generates an HTML file from your PDF.

4. **Recreate the Content:**
   - For the best results and cleanest code, manually recreate your cover letter in HTML using your favorite editor. This gives you full control over the layout and style.

**Tip:** Automated conversions often produce messy HTML. For a professional result (especially for a cover letter or resume), manual recreation is usually best.

**Q: How do I embed PDFs in HTML pages, and what's the best way to manage resume/cover letter files?**

**A:**

1. **Embedding PDFs in HTML:**
   - Use the `<embed>` tag to display PDFs directly on a web page:
     ```html
     <embed src="assets/document.pdf" type="application/pdf">
     ```
   - You can style it with CSS to control width and height:
     ```css
     embed { width: 100%; height: 100vh; }
     ```
   - This allows recruiters and visitors to view your documents directly in the browser and easily print or download them.

2. **Managing Document Files:**
   - Keep your resume and cover letter PDFs in a dedicated folder (e.g., `assets/`).
   - Use clear, consistent file naming (e.g., `LatoyaCarverResume.pdf`, `LatoyaCarverCoverLetter.pdf`).
   - Make sure the HTML files reference the exact file names (case-sensitive on web servers).
   - Always test your file paths locally and after deployment to ensure links work.

3. **Why PDFs are Better Than Images for Documents:**
   - **Professional:** PDFs preserve exact formatting and design from Canva or other design tools.
   - **Printer-friendly:** Recruiters can easily print without formatting issues.
   - **Searchable:** Text in PDFs can be selected and copied.
   - **Shareable:** Easy for recruiters to download and forward to others.

4. **Common Pitfalls to Avoid:**
   - Don't rename files without updating the HTML links (e.g., renaming a .png to .pdf doesn't convert the file).
   - Double-check file names match exactly in both your folder and HTML code.
   - Test all links on your live site after deployment to catch missing files early.

**Key Takeaway:** Embedding PDFs is the professional way to display resumes and cover letters on your portfolio—it looks better than images and is more useful for recruiters.
